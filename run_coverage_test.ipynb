# Get started

Download CTX index file from
https://ode.rsl.wustl.edu/mars/datafile/derived_products/coverageshapefiles/mars/mro/ctx/edr/mars_mro_ctx_edr_c0a.zip
and unpack here

```python
from coverage import Coverage
# Path to downloaded index file
path = './mars_mro_ctx_edr_c0a/mars_mro_ctx_edr_c0a.shp'
```

Initialize the coverage object. This may take a minute to run.
The max_radius_km parameter allows to retain only images within a certain radius around InSight, which will later speed up things.


```python
cover = Coverage(fnam_shape=path, max_radius_km=3000)
```

Next, plot the images in an annulus around InSight

```python
cover.in_annulus(radius_outer_km=1200,
                 radius_inner_km=1000).plot(ec='k')
```

Next, plot only images in an annulus after a given time

```python
cover.in_annulus(radius_outer_km=1200,
                 radius_inner_km=1000,
                 time='2020-01-01').plot(ec='k')
```

Finally, compute the ratio of imaged after a given date vs whole area of annulus.

In my understanding, the resulting warning can be ignored, even though I am still checking...

```python
area = cover.area_images_in_annulus_time(radius_outer_km=1200,
                 radius_inner_km=1000,
                 time='2021-01-01')
ratio = cover.rel_imaged_area_in_annulus(radius_outer_km=1200,
                 radius_inner_km=1000,
                 time='2021-01-01')
print(f'Area: {area:8.2e} km2, which is {ratio*100:4.2f}% of total annulus')


area = cover.area_images_in_annulus_time(radius_outer_km=400,
                 radius_inner_km=200,
                 time='2021-01-01')
ratio = cover.rel_imaged_area_in_annulus(radius_outer_km=400,
                 radius_inner_km=200,
                 time='2021-01-01')
print(f'Area: {area:8.2e} km2, which is {ratio*100:4.2f}% of total annulus')
```
